```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import pynance as pn
import ta

#load test
df = pd.read_csv("../data/raw/AAPL.csv")

df.head()
#data cleaning
df.info()

df.isnull().sum()
#convert date
df["Date"] = pd.to_datetime(df["Date"])

df = df.sort_values("Date")
#enurmes numeric colomen
cols = ["Open", "High", "Low", "Close", "Volume"]

df[cols] = df[cols].apply(pd.to_numeric)
#handling missing
df = df.dropna()
#moving average sma
df["SMA_20"] = ta.trend.sma_indicator(
    close=df["Close"],
    window=20
)

df["SMA_50"] = ta.trend.sma_indicator(
    close=df["Close"],
    window=50
)
#ema
df["EMA_20"] = ta.trend.ema_indicator(
    close=df["Close"],
    window=20
)
#rsi
df["RSI"] = ta.momentum.rsi(
    close=df["Close"],
    window=14
)
#macd
macd = ta.trend.MACD(close=df["Close"])

df["MACD"] = macd.macd()
df["MACD_signal"] = macd.macd_signal()
df["MACD_diff"] = macd.macd_diff()
#Visualization Price + Moving Average
plt.figure(figsize=(14,6))

plt.plot(df["Date"], df["Close"], label="Close")
plt.plot(df["Date"], df["SMA_20"], label="SMA 20")
plt.plot(df["Date"], df["SMA_50"], label="SMA 50")
plt.plot(df["Date"], df["EMA_20"], label="EMA 20")

plt.title("Stock Price with Moving Averages")
plt.legend()
plt.show()
#RSI polt
plt.figure(figsize=(14,4))

plt.plot(df["Date"], df["RSI"], label="RSI")

plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')

plt.title("RSI Indicator")
plt.legend()

plt.show()
#macd plot
plt.figure(figsize=(14,5))

plt.plot(df["Date"], df["MACD"], label="MACD")
plt.plot(df["Date"], df["MACD_signal"], label="Signal")

plt.bar(df["Date"], df["MACD_diff"], label="Histogram")

plt.legend()
plt.title("MACD")

plt.show()
#PyNance Metrics
df["returns"] = df["Close"].pct_change()

volatility = df["returns"].std()

print(volatility)
```
